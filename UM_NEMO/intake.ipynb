```python
import warnings

import iris
import intake
```

```python
def load_callback(cube, field, filename):
    """
    An Iris load callback function to delete a few metadata items that will
    prevent the files from being merged into a single cube.
    """
    attributes_to_remove = [
        "creation_date",
        "history",
        "tracking_id"
    ]
    for attr_name in attributes_to_remove:
        del(cube.attributes[attr_name])
```

```python
dc = intake.open_esm_datastore("https://raw.githubusercontent.com/eerie-project/intake_catalogues/main/jasmin/jasmin-catalogue.json")
```

```python
dc_subset = dc.search(
    source_id="HadGEM3-GC5-EERIE-N640-ORCA12",
    experiment_id="eerie-piControl",
    table_id="Amon",
    variable_id="tas"
)
dc_subset
```

```python
dc_subset_paths = [path_name for path_name in dc_subset.df.path]
with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    cubes = iris.load(dc_subset_paths, callback=load_callback)
tas = cubes.concatenate_cube()
```
